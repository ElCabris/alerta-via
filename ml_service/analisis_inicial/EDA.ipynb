```python
# Crear un mapeo de barrio -> número entero único
barrio_map = {barrio: idx for idx, barrio in enumerate(data['codigo_barrio'].str.replace('#','', regex=False).unique())}

data['codigo_barrio_num'] = data['codigo_barrio'].str.replace('#','', regex=False).map(barrio_map)

print(data[['codigo_barrio', 'codigo_barrio_num']].head())
```

```python
data['codigo_barrio_num'].nunique()
```

```python
data['codigo_comuna'] = pd.to_numeric(data['codigo_comuna'], errors='coerce')

print(data['codigo_comuna'].tail())
print(data['codigo_comuna'].dtype)
```

```python
del data['nombre_barrio']
```

```python
from sklearn.preprocessing import LabelEncoder

# Lista de columnas a codificar
cols_to_encode = ['modalidad', 'lugar', 'sede_receptora', 'bien', 'categoria_bien']

# Diccionario para guardar las clases de cada columna (para interpretación futura)
label_encoders = {}

for col in cols_to_encode:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    label_encoders[col] = le.classes_
    print(f"✅ Columna '{col}' codificada. Número de categorías: {len(le.classes_)}")

print("\nTipos de datos después de encoding:")
print(data[cols_to_encode].dtypes)
```

```python
# Columnas para One-Hot Encoding
cols_to_one_hot = ['sexo', 'estado_civil', 'medio_transporte', 'arma_medio', 'grupo_bien']

data = pd.get_dummies(data, columns=cols_to_one_hot, drop_first=False, dtype=int)

print("✅ One-Hot Encoding aplicado con 0 y 1.")
print(f"📊 Nuevas dimensiones del dataset: {data.shape}")
print(f"🔑 Columnas creadas: {', '.join([c for c in data.columns if any(col in c for col in cols_to_one_hot)])[:200]}...")
```

```python
data.info()
```

```python
data['year'] = data['fecha_hecho'].dt.year
data['month'] = data['fecha_hecho'].dt.month
data['day'] = data['fecha_hecho'].dt.day
data['dayofweek'] = data['fecha_hecho'].dt.dayofweek
data['hour'] = data['fecha_hecho'].dt.hour
data['is_weekend'] = data['fecha_hecho'].dt.dayofweek.isin([5, 6]).astype(int)

data = data.drop(columns=['fecha_hecho'])
```
